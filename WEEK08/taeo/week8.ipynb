3. generator μ λ„λ μ΄ν„°
β–¶  iteratorλ¥Ό μƒμ„±ν•λ” ν•¨μμ΄κ³ , generatorλ΅ μƒμ„±ν• κ°μ²΄λ” iteratorμ™€ λ§μ°¬κ°€μ§€λ΅ next()ν•¨μλ¥Ό νΈμ¶ν•΄ λ‹¤μ κ°’ μ–»κΈ° κ°€λ¥ν•λ‹¤. μ΄ν„°λ μ΄ν„°μ™€λ” λ‹¤λ¥΄κ² yield λΌλ” λ¬Έλ²•μ„ μ‚¬μ©ν•λ―€λ΅ ν•΄λ‹Ή ν•¨μ(generator)κ°€ μΆ…λ£ν•μ§€ μ•κ³ , κ·Έλ€λ΅ μ μ§€λλ‹¤. 
yield λ” μΌλ° ν•¨μμ returnκ³Ό κ°™μ΄ κ°’μ„ λ°ν™ ν•μ§€λ§, returnκ³Ό λ‹¤λ¥΄κ² ν•΄λ‹Ή ν•¨μ(generator)κ°€ μΆ…λ£ν•μ§€ μ•κ³ , κ·Έλ€λ΅ μ μ§€λκ³  λ‹¤μ μμ„μ μ λ„λ μ΄ν„°(Generator)κ°€ νΈμ¶λλ©΄, λ©μ¶”μ—λ yield μλ¦¬μ—μ„ λ‹¤μ‹ ν•¨μκ°€ λ™μ‘ν•κ² λλ” μ›λ¦¬μ΄λ‹¤.
μ¦‰,μ λ„λ μ΄ν„°(Generator) β†’ next(μ λ„λ μ΄ν„°) β†’ μ λ„λ μ΄ν„° ν•¨μ μ‹¤ν–‰ β†’ yieldλ¥Ό λ§λ‚ next(μ λ„λ μ΄ν„°κ°€) νΈμ¶λ κ³³μΌλ΅ κ°’μ„ λ°ν™ (μ λ„λ μ΄ν„° μΆ…λ£ β) β†’ λ‹¤μ‹ next(μ λ„λ μ΄ν„°) β†’ λ©μ¶”μ—λ yieldλ¶€ν„° μ¬μ‹¤ν–‰

```python
def my_generator():
    print("μ²« λ²μ§Έ κ°’ μƒμ„±")
    yield 1
    print("λ‘ λ²μ§Έ κ°’ μƒμ„±")
    yield 2
    print("μ„Έ λ²μ§Έ κ°’ μƒμ„±")
    yield 3

# μ‚¬μ©
gen = my_generator()

print(next(gen))  # μ²« λ²μ§Έ κ°’ μƒμ„±, 1 μ¶λ ¥
print(next(gen))  # λ‘ λ²μ§Έ κ°’ μƒμ„±, 2 μ¶λ ¥
print(next(gen))  # μ„Έ λ²μ§Έ κ°’ μƒμ„±, 3 μ¶λ ¥
# print(next(gen))  μ„ μ‹¤ν–‰ν•λ©΄ StopIteration λ°μƒ
```

```python
#ν΄λμ¤ μ–΄λ…Έν…μ΄μ…
class Student:
    def __init__(self, name:str, age:int):
        self.name=name
        self.age=age
    def __str__(self)->str:
        return 'μ΄λ¦„: ' + self.name + 'λ‚μ΄' + str(self.age)
# print() ν•¨μλ§ μ“΄λ‹¤λ©΄ λ¦¬ν„΄κ°’μ€ NoneμΌλ΅ λ‚μ¤μ§€λ§, λ¦¬ν„΄κ°’μ„ μ„¤μ •ν•΄ λ©”λ¨λ¦¬
#κ°’μ„ λ„£μμΌλ΅μ¨ μ¤λ²„λΌμ΄λ”© λ κ²ƒ

def a(s:Student) ->None:
    print(f'μ΄λ¦„:{s.name}, λ‚μ΄:{s.age} π–')

student=Student('peter', 25)    # classλ¬Έ->__init__->__str__ μ¤λ²„λΌμ΄λ”©
print(student)
a(student)
```
